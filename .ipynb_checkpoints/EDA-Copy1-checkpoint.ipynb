```python
# Load and read data
file = "Resources/mortgage_rates.csv"
df2=pd.read_csv(file,parse_dates=True)
df2.dtypes
```

```python
df2['Date'] = pd.to_datetime(df2['Date'])
```

```python
Final_db=pd.merge(Final, df2, on = "Date", how = "left")
Final_db
```

```python
Final_db.isnull().sum()
```

```python
# Fill the missing values using mean method
missing_col = ['Rate','Townhouse','Apartment']
#Technique 1: Using mean to impute the missing values
for i in missing_col:
 Final_db.loc[Final_db.loc[:,i].isnull(),i]=Final_db.loc[:,i].mean()
```

```python
g= sns.relplot(x='Date', y='Composite', hue='Region', kind='line', data=Final_db);
g.fig.set_figwidth(18)
g.fig.set_figheight(10)
plt.legend(bbox_to_anchor=(1.02, 0.15), loc='upper left', title='Region', borderaxespad=0)
plt.xlabel('Date', fontsize=16);
plt.ylabel('House Price', fontsize=16);
plt.title('Historical House Price by Region', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=14)
```

```python
# Create our features
X = Final_db.drop(["Composite","Date","Single_family","One_storey","Two_storey","Townhouse","Apartment"],axis=1)
X = pd.get_dummies(X)

# Create our target
y = Final_db.Composite
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# print the intercept
print(model.intercept_)
```

```python
coeff_parameter = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_parameter
```

```python
predictions = model.predict(X_test)
predictions
```

```python
sns.regplot(y_test,predictions)
```
