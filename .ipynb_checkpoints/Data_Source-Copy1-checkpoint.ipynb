```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
```

```python
# Load and read data
final_data=["Data_Source/ONTARIO.csv","Data_Source/QUEBEC.csv","Data_Source/BC.csv","Data_Source/ALBERTA.csv",
           "Data_Source/SASKATCHEWAN.csv","Data_Source/NEWFOUNDLAND.csv"]
dfs=[pd.read_csv(f) for f in final_data]
dfs[0]
```

```python
dfs[0] =dfs[0].drop(columns=['Composite_HPI_SA','Single_Family_HPI_SA','One_Storey_HPI_SA','Two_Storey_HPI_SA','Townhouse_HPI_SA','Apartment_HPI_SA'])
dfs[0]=dfs[0].set_index("Date")
dfs[0]
```

```python
col={'Composite_Benchmark_SA':'Composite','Single_Family_Benchmark_SA':'Single_family',
     'One_Storey_Benchmark_SA':'One_storey','Two_Storey_Benchmark_SA':'Two_storey',
     'Townhouse_Benchmark_SA':'Townhouse','Apartment_Benchmark_SA':'Apartment'}
```

```python
dfs[0].rename(columns=col,inplace=True)
```

```python
dfs[0]=dfs[0].stack()
dfs[0]
```

```python
ON=pd.DataFrame(dfs[0])
ON['Region']='Ontario'
```

```python
ON.to_csv("Final_data/Ontario.csv")
```

```python
dfs[1] =dfs[1].drop(columns=['Composite_HPI_SA','Single_Family_HPI_SA','One_Storey_HPI_SA','Two_Storey_HPI_SA','Townhouse_HPI_SA','Apartment_HPI_SA'])
dfs[1]=dfs[1].set_index("Date")
dfs[1]
```

```python
dfs[1].rename(columns=col,inplace=True)
dfs[1]=dfs[1].stack()
dfs[1]
```

```python
QB=pd.DataFrame(dfs[1])
QB['Region']='Quebec'
QB.to_csv("Final_data/Quebec.csv")
```

```python
dfs[2] =dfs[2].drop(columns=['Composite_HPI_SA','Single_Family_HPI_SA','One_Storey_HPI_SA','Two_Storey_HPI_SA','Townhouse_HPI_SA','Apartment_HPI_SA'])
dfs[2]=dfs[2].set_index("Date")
dfs[2].rename(columns=col,inplace=True)
dfs[2]=dfs[2].stack()
BC=pd.DataFrame(dfs[2])
BC['Region']='British_Columbia'
BC.to_csv("Final_data/British_Columbia.csv")
```

```python
dfs[3] =dfs[3].drop(columns=['Composite_HPI_SA','Single_Family_HPI_SA','One_Storey_HPI_SA','Two_Storey_HPI_SA','Townhouse_HPI_SA','Apartment_HPI_SA'])
dfs[3]=dfs[3].set_index("Date")
dfs[3].rename(columns=col,inplace=True)
dfs[3]=dfs[3].stack()
AB=pd.DataFrame(dfs[3])
AB['Region']='Alberta'
AB.to_csv("Final_data/Alberta.csv")
```

```python
dfs[4] =dfs[4].drop(columns=['Composite_HPI_SA','Single_Family_HPI_SA','One_Storey_HPI_SA','Two_Storey_HPI_SA','Townhouse_HPI_SA','Apartment_HPI_SA'])
dfs[4]=dfs[4].set_index("Date")
dfs[4].rename(columns=col,inplace=True)
dfs[4]=dfs[4].stack()
SK=pd.DataFrame(dfs[4])
SK['Region']='Saskatchewan'
SK.to_csv("Final_data/Saskatchewan.csv")
```

```python
dfs[5] =dfs[5].drop(columns=['Composite_HPI_SA','Single_Family_HPI_SA','One_Storey_HPI_SA','Two_Storey_HPI_SA','Townhouse_HPI_SA','Apartment_HPI_SA'])
dfs[5]=dfs[5].set_index("Date")
dfs[5].rename(columns=col,inplace=True)
dfs[5]=dfs[5].stack()
NF=pd.DataFrame(dfs[5])
NF['Region']='Newfoundland'
NF.to_csv("Final_data/Newfoundland.csv")
```

```python
Price_df=pd.concat([ON,QB,BC,AB,SK,NF],axis=0)
Price_df.head()
```

```python
Price_df.reset_index(inplace=True)
row=len(Price_df.index)
Price_df=Price_df.iloc[0:row]
Price_df
```

```python
Price_df=Price_df.rename(columns = {"Date" : "Date","level_1":"Features",
                                   "":"Price","Region":"Region"})
Price_df
```

```python
Price_df.to_csv("Final_data/Price.csv")
```
