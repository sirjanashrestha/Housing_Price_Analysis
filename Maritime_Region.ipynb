```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
```

```python
# Load and read csv
file_to_load = "Maritime_regions.csv"
maritime_df = pd.read_csv(file_to_load)
maritime_df
```

```python
maritime_df = maritime_df.drop(columns=["Composite_HPI_SA", 'Single_Family_HPI_SA', 'One_Storey_HPI_SA','Two_Storey_HPI_SA','Townhouse_HPI_SA','Apartment_HPI_SA', "Region"], axis = 1)
maritime_df = maritime_df.set_index("Date") 
maritime_df
```

```python
col={'Composite_Benchmark_SA':'Composite','Single_Family_Benchmark_SA':'Single_family',
     'One_Storey_Benchmark_SA':'One_storey','Two_Storey_Benchmark_SA':'Two_storey',
     'Townhouse_Benchmark_SA':'Townhouse','Apartment_Benchmark_SA':'Apartment'}
```

```python
maritime_df.rename(columns=col,inplace=True)
```

```python
maritime_df=maritime_df.stack()
maritime_df
```

```python
Maritime=pd.DataFrame(maritime_df)
Maritime["Region"] = "Maritime_Regions"
Maritime.to_csv("Maritime.csv")
```
